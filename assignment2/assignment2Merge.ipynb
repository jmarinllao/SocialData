```python
import warnings

from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score


from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, ShuffleSplit
```


## Part 2A: Random forest binary classification

**1.** Using instructions and material from Week 7, **build a `random forest` classifier to distinguish between two types (you choose) of crime using on spatio-temporal (where/when) features** of data describing the two crimes. When you're done, you should be able to give the classifier a place and a time, and it should tell you which of the two types of crime happened there.

> **Data Prep**

> Time measures

```python
crime_data["Hour"]= pd.to_datetime(crime_data["Time"], format='%H:%M').dt.hour
crime_data["Date"]= pd.to_datetime(crime_data["Date"], format='%m/%d/%Y')
crime_data["Month"] = crime_data['Date'].dt.month
crime_data["Day"] = crime_data['Date'].dt.day
```

> Crime categories analyzed

```python
type1="VEHICLE THEFT" #influenced by weather conditions
type2="FRAUD" #not influenced
```

> Take the same amount of samples and generate dataframe containing dataset for classification

```python
df1=crime_data[crime_data["Category"]==type1].sample(12000)[['Hour', 'Day', 'Month', 'PdDistrict','Category','Date']]
df2=crime_data[crime_data["Category"]==type2].sample(12000)[['Hour', 'Day', 'Month', 'PdDistrict','Category','Date']]
df=pd.concat([df1,df2])
```

> Categorical variables encoding

```python
df['PdDistrict']=df['PdDistrict'].astype('category')
df['PdDistrict_cat']=df['PdDistrict'].cat.codes
```

> **Train/test split** -> 70/30

```python
msk = np.random.rand(len(df)) < 0.7
df_train = df[msk]
df_test = df[~msk]
```

> **Train with `grid CV`**

```python
ranfor_params = {'fit__n_estimators' : range(20,200,5) , 'fit__max_depth' : range(1,6) , 'fit__max_features' : ["sqrt","log2"]}

pipe = Pipeline( [('fit',RandomForestClassifier(bootstrap=True, oob_score=True,random_state=0))] )

with warnings.catch_warnings(): # to disable all the convergence warnings from GridSearchCV
    warnings.simplefilter("ignore")
    randomForest = GridSearchCV(pipe, 
                                param_grid=ranfor_params,
                                scoring='accuracy',
                                cv = 5, verbose=0, n_jobs=-1).fit( np.array(df_train[['Hour', 'Day', 'Month', 'PdDistrict_cat']]), df_train['Category']).best_estimator_
```

> **Performance metrics**

```python
with warnings.catch_warnings(): # to disable all the convergence warnings from GridSearchCV
    warnings.simplefilter("ignore")
    #print(randomForest)
    print("Accuracy from CV results on train set: "+str(randomForest.score( np.array(df_train[['Hour', 'Day', 'Month', 'PdDistrict_cat']]), df_train['Category'])))

    #predict using best parameters
    predictions=randomForest.predict(np.array(df_test[['Hour', 'Day', 'Month', 'PdDistrict_cat']]))

    #compute evaluation measures
    accuracy = accuracy_score(df_test['Category'],predictions)
    print("Accuracy on the test set: "+str(accuracy))
```

1. **Explain about your choices for training/test data, features and encoding.** 

> * Data is splited in 80% *train* and 20% *test*. 
> * Time varibales were differenciated as much as possible in order to have more predictors to deal with, since we have a lot of samples, the curse of dimensionality is not a main issue in this problem. 
> * Categorical variables were treated with one-hot encoding, which might add many misleading features. Such encoding is more appropriate to be implemented with regularization models. 
> * `GridSearchCV` was employed as an "Exhaustive search over specified parameter values for an estimator." \[1\]  It allows us to perform parameter tuning on the model, thus finding the optimal ones. 
> * Random forests are ensemble methods known for rarely *overfitting* with many trees and proper parameter values \[2\]. Then, adding CV gave us more reliability in the parameter tuning generalization due to the repeated train/validation check. Thus, we can see that the C.V. training accuracy is really close to the Testing accuracy, this suggest us that we have a good generalization. On the other hand having both values above the random value we expect our model to not underfit as well. 

2. **Discuss the model performance. Report accuracy.**

> As stated above,we see that the performance is better than random. Accuracy is the number of correct predictions made over all the predictions. We can observe a performance on the training set of 67.11% of correct classified elements and 66.80% on the test set. This means that (as expected) we perform a little worse on unseen data, but with our reliable tuning procedure we can expect a performance on more unseen data point similar to that one we had on our test set.

## Part 2B: Info from weather features

Now we add features from weather data to our random forest

Merge databases

```python
weather_data["Hour"]     =  pd.to_datetime(weather_data.index, format="%Y-%m-%dT%H:%M:%S.%fZ").hour
weather_data["Date"]     =  pd.to_datetime(weather_data.index, format="%Y-%m-%dT%H:%M:%S.%fZ").date
weather_data["DateHour"] =  weather_data["Date"].astype(str)+"-"+weather_data["Hour"].astype(str)
df1["DateHour"]     =  df1["Date"].astype(str)+"-"+df1["Hour"].astype(str)
df2["DateHour"]     =  df2["Date"].astype(str)+"-"+df2["Hour"].astype(str)

df1.set_index(["DateHour"],inplace= True)
df2.set_index(["DateHour"],inplace= True)
weather_data.set_index(["DateHour"],inplace= True)

merge1=pd.merge(df1,weather_data, how='inner', left_index=True, right_index=True)
merge1=merge1.dropna()
merge2=pd.merge(df2,weather_data, how='inner', left_index=True, right_index=True)
merge2=merge2.dropna()
```

Categorical codifications

```python
df=pd.concat([merge1.sample(900),merge2.sample(900)])#due to the smaller csv bug

df['weather']=df['weather'].astype('category')
df['Weather_cat']=df['weather'].cat.codes


df['PdDistrict']=df['PdDistrict'].astype('category')
df['PdDistrict_cat']=df['PdDistrict'].cat.codes
```

**Train/test split** -> 70/30

```python
msk = np.random.rand(len(df)) < 0.7
df_train = df[msk]
df_test = df[~msk]
```

**Train with `grid CV`**

```python
ranfor_params = {'fit__n_estimators' : range(20,200,5) , 'fit__max_depth' : range(1,6) , 'fit__max_features' : ["sqrt","log2"]}

pipe = Pipeline( [('fit',RandomForestClassifier(bootstrap=True, oob_score=True,random_state=0))] )


with warnings.catch_warnings(): # to disable all the convergence warnings from GridSearchCV
    warnings.simplefilter("ignore")
    randomForest = GridSearchCV(pipe, 
                                param_grid=ranfor_params,
                                scoring='accuracy',
                                cv = 7, verbose=0, n_jobs=-1).fit(np.array(df_train[['Hour_x', 'Day', 'Month', 'PdDistrict_cat','Weather_cat','temperature']]), df_train['Category']).best_estimator_
```

**Performance metrics**

```python
with warnings.catch_warnings(): # to disable all the convergence warnings from GridSearchCV
    warnings.simplefilter("ignore")
    #print(randomForest)
    print("Accuracy from CV results on train set: "+str(randomForest.score( np.array(df_train[['Hour_x', 'Day', 'Month', 'PdDistrict_cat','Weather_cat','temperature']]), df_train['Category'])))

    #predict using best parameters
    predictions=randomForest.predict(np.array(df_test[['Hour_x', 'Day', 'Month', 'PdDistrict_cat','Weather_cat','temperature']]))

    #compute evaluation measures
    accuracy = accuracy_score(df_test['Category'],predictions)
    print("Accuracy on the test set: "+str(accuracy))
```

## Part 3A:. Data Prep

A key step is to set up the data rightfor the dataviz, following the workflow:

**1.** Take the data for the **period 2010-2018** and group it by hour-of-the-day

> Transform *Date* column into a time format.

```python
crime_data["Date"]=pd.to_datetime(crime_data["Date"], format='%m/%d/%Y')
```

> Select all the entries between the indicated period.

```python
crime_data_10_18 = crime_data[crime_data["Date"].between('01/01/2010', '31/12/2018')]
```

> Group data by hour-of-the-day.

```python
crime_data["Hour"]= pd.to_datetime(crime_data["Time"], format='%H:%M').dt.hour 
crime_data["HourInt"]= [f'{str(h)}h-{str(h+1)}h' for h in crime_data["Hour"]]

crime_count_by_hour_and_category=crime_data[crime_data["Category"].isin(focuscrimes)
                                           ].groupby(["Category", "Hour", "HourInt"]).count()["IncidntNum"].reset_index()
```

**2.** **Normalize data** to be able to easily compare how the distribution of crimes differ from each other, not absolute numbers.
To normalise data for within a crime category we devide the count for each hour by the total number of this crime type.

> Incident number **total sum by category**.

```python
crime_count_by_category=crime_data[crime_data["Category"].isin(focuscrimes)
                                           ].groupby(["Category"]).count()["IncidntNum"].reset_index()
```

> **Divide** each entry in the table grouped **by hour-of-the-day** by its total incident number **by category** (one-line `lambda` *functional programing*), storing it into a new column `NormalizedIncidntNum`.

```python
crime_count_by_hour_and_category['NormalizedIncidntNum'] = crime_count_by_hour_and_category.apply(lambda x: x.IncidntNum/crime_count_by_category[crime_count_by_category["Category"]==x.Category].IncidntNum.values[0], axis=1)
```

> Reshape the normalized data `NormalizedIncidntNum` into a table with **`crime-type` as columns and `hours` as rows**.

```python
crime_count_by_hour_and_category = crime_count_by_hour_and_category.pivot(index=crime_count_by_hour_and_category.Hour, columns='Category')['NormalizedIncidntNum']
```

## Part 2B:. Bokeh Visualization

```python
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.plotting import figure, output_file, show
from bokeh.palettes import turbo
```

Bokeh view into the *JupyterNotebook*:

```python
from bokeh.io import push_notebook, show, output_notebook
output_notebook()
```

We follow the **Bokeh guide for categorical data**, in order to plot our ready preprocessed data:

> **1.** Convert our `Pandas Dataframe` to `Bokeh ColumnDataSource`: 

```python
source = ColumnDataSource(crime_count_by_hour_and_category)
```

> **2.** Create an empty figure, in which we will input our data later on: 

```python
hours = [str(e) for e in list(crime_count_by_hour_and_category.index)]
```

```python
p = figure(x_range=FactorRange(factors=hours), plot_height=500, width=1000, title="Crimes per hour",
           toolbar_location=None, tools="", x_axis_label='Hour of the day', y_axis_label='Relative Fequency')
```

> **3.** Add the bars, with `vbar`: 

```python
bar ={} # to store vbars
palette = turbo(len(focuscrimes)) # selected palette of the size of types of crimes

for indx,i in enumerate(focuscrimes):
    bar[i] = p.vbar(x='Hour',  top=i, source= source, 
                 ### create a vbar for each focuscrime
                 legend_label=i,  color = palette[indx], muted_alpha=0.1, muted =True) 

    p.legend.click_policy="mute" ### assigns the click policy
```

> **4.** Set **legend** out-right of the plot: 

```python
p.legend.location = "top_left"
p.add_layout(p.legend[0], 'right')
```

> **5.** Display the interactive figure, within  the `jupyter notebook`: 

```python
show(p, notebook_handle=True)
```
